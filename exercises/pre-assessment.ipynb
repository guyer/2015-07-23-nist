# NumPy and Matplotlib

```python
import numpy as np
from matplotlib import pyplot as plt
import csv
```

```python
with open('NIST Pre-Workshop Assessment Form (Responses) - Form Responses scrub.csv', 'Urb') as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='"')
    preassessment_np = np.array([line for line in reader])
```

```python
for l in enumerate(preassessment_np[0]):
    print l
```

```python
import textwrap
def assess_histo_np_mpl(assessment, query, labels):
    responses = assessment[1:, assessment[0] == query][..., 0]
    frequencies = [(responses == label).sum() for label in labels]
    
    pos = np.arange(len(labels))
    width = 1.0

    ax = plt.axes()
    ax.set_title(textwrap.fill(query, width=50))
    ax.set_yticks(pos + (width / 2))
    
    labels = [label.replace('\xc2\xa0', ' ') for label in labels]
    ax.set_yticklabels(labels) 

    plt.barh(pos, frequencies, width, color='r')
    plt.show()
```

## shell

```python
assess_histo_np_mpl(assessment=preassessment_np,
                    query='How familiar do you think you are with the command line?',
                    labels=['I am not familiar with the command line.', 
                            'I am familiar with only the term "command line."', 
                            'I am familiar with the command line but have never used it.', 
                            'I am familiar with the command line because I have or am using it.'])
```

```python
assess_histo_np_mpl(assessment=preassessment_np,
                    query='How would you solve this problem: A directory contains 1000 text files. Create a list of all files that contain the word "Drosophila" and save the result to a file called results.txt.',
                    labels=['I could not\xc2\xa0create this list.', 
                            'I would create this list using "Find in Files" and "copy and paste."', 
                            'I would create this list using basic command line programs.',
                            'I would create this list using a pipeline of command line programs.'])
```

## programming

```python
def language_histo_np_mpl(assessment, query, languages):
    responses = assessment[1:, assessment[0] == query][..., 0]
    
    frequencies = [np.sum([language.lower() in response.lower().split(', ') 
                           for response in responses])
                   for language in languages]
    
    pos = np.arange(len(languages))
    width = 1.0

    ax = plt.axes()
    ax.set_title(textwrap.fill(query, width=50))
    ax.set_yticks(pos + (width / 2))
    ax.set_yticklabels(languages)

    plt.barh(pos, frequencies, width, color='r')
    plt.show()
```

```python
language_histo_np_mpl(assessment=preassessment_np,
                      query='With which programming languages, if any, could you '
                      'write a program from scratch which imports some data and '
                      'calculates mean and standard deviation of that data?', 
                      languages=["none", "C", "C++", "Perl", "MATLAB", "Python", "R", "Java"])
```

# pandas

```python
import pandas as pd
```

```python
preassessment_pd = pd.read_csv("NIST Pre-Workshop Assessment Form (Responses) - Form Responses scrub.csv")
```

Plotting a histogram of the assessment answers is considerably easier than when using NumPy and Matplotlib directly. We simply provide the name of the column and ask for its `value_counts()`, which automatically tallies all of the responses.

```python
query = 'How familiar are you with unit testing and code coverage?'
preassessment_pd[query].value_counts().plot(kind='barh', title=query)
```

With one line of pandas code to import the data and another line to process and display, we've accomplished what it took at least 16 lines of NumPy and Matplotlib code to achieve. 

A few things are not optimal, however. `value_counts()` cannot tally answers it doesn't know about (we'd like to know if nobody chose a particular answer) and it automatically orders from most responses to least. We can remedy both problems with a call to `reindex()`.

```python
labels = ['I am not familiar with\xc2\xa0unit testing or code coverage.', 
          'I am familiar only with the terms "unit testing" and "code coverage".', 
          'I am familiar with\xc2\xa0unit testing or code coverage\xc2\xa0but have never used it.', 
          "I am familiar with\xc2\xa0unit testing or code coverage\xc2\xa0because I have used or am using them."]
preassessment_pd[query].value_counts().reindex(index=labels, fill_value=0).plot(kind='barh', title=query)
```

Finally, we wrap the text of the title to keep things looking "nice".

```python
import textwrap
preassessment_pd[query].value_counts().reindex(index=labels, fill_value=0).plot(kind='barh', title=textwrap.fill(query, 50))
```

Even though it's only one line of code, we encapsulate this into a Python function to keep things more readable

```python
import textwrap
def assess_histo_pd(assessment, query, labels):
    counts = assessment[query].value_counts()
    counts = counts.reindex(index=labels, fill_value=0)
    counts.plot(kind='barh', title=textwrap.fill(query, 50))
```

```python
assess_histo_pd(preassessment_pd,
                query='How familiar are you with unit testing and code coverage?',
                labels=['I am not familiar with\xc2\xa0unit testing or code coverage.', 
                        'I am familiar only with the terms "unit testing" and "code coverage".', 
                        'I am familiar with\xc2\xa0unit testing or code coverage\xc2\xa0but have never used it.', 
                        "I am familiar with\xc2\xa0unit testing or code coverage\xc2\xa0because I have used or am using them."])
```

If desired, you could fiddle further to make the plot look identical to the NumPy/Matplotlib version above.

## programming

```python
query = 'With which programming languages, if any, could you ' \
        'write a program from scratch which imports some data and ' \
        'calculates mean and standard deviation of that data?'
        
# we cast everything to lower case to ensure that we count all variant spellings
languages = preassessment_pd[query].apply(lambda x: pd.Series(x.lower().split(', ')))
counts = languages.stack().value_counts()

counts.plot(kind='barh', title=textwrap.fill(query, 50))
```

This gives a much more complete picture than the NumPy version above, where we had to supply an explicit list of languages. There are a few anomalies due to the formatting of the responses, but they're not worth trying to correct for this quick overview.
